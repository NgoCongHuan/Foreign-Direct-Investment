```python
import pandas as pd
```

```python
# Load tập dữ liệu
df = pd.read_csv('/content/drive/MyDrive/Foreign-Direct-Investment/data/fdi_country_partners_en.csv',index_col='Order')
```

```python
# In ra 5 dòng đầu tiên
df.head(5)
```

```python
# Thống kê các cột có trong DataFrame
print("Column before modification:")
for x, column in enumerate(df.columns):
    print("Column {}: {}".format(x, column))

print("------------------------------")
print("Modification processing...")
# Loại bỏ khoảng trắng ở đầu và cuối mỗi tên cột
df.columns = df.columns.str.strip()

# Đổi ký tự " \n" thành " "
df.columns = df.columns.str.replace(" \n"," ")

# Đổi ký tự "\n" thành ""
df.columns = df.columns.str.replace("\n"," ")
print("------------------------------")

# Kiểm tra lại các cột có trong DataFrame
print("Column after modification:")
for x, column in enumerate(df.columns):
    print("Column {}: {}".format(x, column))
```

```python
columns_mapping = {'Country': 'Đối tác',
                  'Number of new projects': 'Số dự án cấp mới',
                  'Newly registered capital (million USD)': 'Vốn đăng ký cấp mới (triệu USD)',
                  'Adjusted project number': 'Số lượt dự án điều chỉnh',
                  'Adjusted capital (million USD)': 'Vốn đăng ký điều chỉnh (triệu USD)',
                  'Number of times of capital contribution to buy shares': 'Số lượt góp vốn mua cổ phần',
                  'Value of capital contribution, share purchase (million USD)': 'Giá trị góp vốn, mua cổ phần (triệu USD)',
                  'Year': 'Năm'}
```

```python
df.rename(columns=columns_mapping, inplace=True)
```

```python
df.head()
```

```python
# Khởi tạo mảng chứa các cột cần loại bỏ khoảng trắng trong mỗi giá trị
column_to_strip = df.drop('Năm', axis=1).columns

# Khởi tạo vòng lặp duyệt qua các giá trị trong cột và loại bỏ khoảng trắng dư thừa
for column in column_to_strip:
  df[column] = df[column].str.strip()
```

```python
# Thống kê số lượng các giá trị thiếu ở mỗi cột
print(df.isnull().sum())
print("-> The total number of initial null values is: {}".format(df.isnull().sum().sum()))

print("------------------------------")
print("Fillna processing...")
# Thay thế số giá trị null thành 0
# Thay thế các chuỗi có dạng " -" và "- " thành ""
df = df.replace('-',0)
df = df.fillna(0)
print("------------------------------")

# Kiểm tra lại số giá trị null
print(df.isnull().sum())
print("-> The total number of null values after substitution is: {}".format(df.isnull().sum().sum()))
```

```python
# Khởi tạo mảng chứa các cột cần chuyển đổi kiểu int
column_to_int = ['Số dự án cấp mới', 'Số lượt dự án điều chỉnh', 'Số lượt góp vốn mua cổ phần']

# Khởi tạo vòng lặp để chuyển đổi các cột sang kiểu int
for column in column_to_int:
  df[column] = df[column].replace(',','', regex=True)
  df[column] = df[column].astype(int)
```

```python
# Khởi tạo mảng chứa các cột cần chuyển đổi kiểu float
column_to_float = ['Vốn đăng ký cấp mới (triệu USD)', 'Vốn đăng ký điều chỉnh (triệu USD)', 'Giá trị góp vốn, mua cổ phần (triệu USD)']

# Khởi tạo vòng lặp để chuyển đổi các cột sang kiểu float
for column in column_to_float:

  # Xóa các kí tự đặc biệt
  df[column] = df[column].replace(',','', regex=True)
  df[column] = df[column].replace('\(','',regex=True)
  df[column] = df[column].replace('\)','',regex=True)

  # Chuyển đổi sang kiểu float
  df[column] = df[column].astype(float)
```

```python
import pmdarima as pm

############################## Hàm Thêm dữ liệu ##############################
def fill_data(df):

  # Thống kê các năm và đối tượng
  objects = df.iloc[:,0].unique()
  years = df.iloc[:,-1].unique()

  # Duyệt qua từng đối tượng
  for obj in objects:

    # DataFrame tạm thời chứa thông tin nguồn vốn của đối tượng qua các năm
    df_temp = df[df.iloc[:,0] == obj]

    # Thống kê các năm mà đối tượng này không có dữ liệu
    years_of_obj = [y for y in years if y not in df_temp.iloc[:,-1].values]

    # Thêm dữ liệu vào các năm mà đối tượng không tồn tại
    if years_of_obj:

      for year in years_of_obj:

        data = {}

        data[df.columns[0]] = obj
        data[df.columns[-1]] = year

        for col in df.columns[1:-1]:
          data[col] = 0

        df_new = pd.DataFrame([data])

        df = pd.concat([df,df_new], ignore_index=True)

  return df

############################## Hàm xoá dữ liệu đã thêm ##############################
def remove_fill_data(df):

  df = df[df[df.columns[1:-1]].sum(axis=1) != 0]

  return df

############################## Hàm dự đoán từng đặc trưng ##############################
def predict_fdi(df):

  df = df.sort_values(by='Năm')

  data = df.iloc[:,1].values

  # Khởi tạo mô hình
  model = pm.auto_arima(data, seasonal=False, trace=True)

  n_periods = 3
  forecasts = model.predict(n_periods=n_periods)

  results = forecasts

  return results

############################## Hàm dự đoán tất cả đặc trưng ##############################
def main(df):

  df = fill_data(df)

  for obj in df.iloc[:,0].unique():

    df_temp = df[df.iloc[:,0] == obj]

    data_predict = {}

    for col in df_temp.columns[1:-1]:

      results = predict_fdi(df_temp[[df_temp.columns[0], col, df_temp.columns[-1]]])

      data_predict[col] = results

    data_predict[df_temp.columns[-1]] = [y for y in range(2023,2026)]

    data_predict[df_temp.columns[0]] = [obj for i in range(2023,2026)]

    df_predict = pd.DataFrame(data_predict)

    df = pd.concat([df, df_predict], ignore_index=True)

  df = remove_fill_data(df)

  df = df.reset_index(drop=True)

  return df

df = main(df)
```
